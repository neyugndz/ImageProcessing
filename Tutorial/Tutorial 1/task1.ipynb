## Task 1

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('E:\\USTH\\Image Processing\\Tutoring\\Tutorial 1\\Data\\images2.jpg', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
if img is not None:
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')
    plt.show()
else:
    print("Failed to lead the image. Please check the file path")
```

## Task 2

```python
# Resize image
half = cv2.resize(img_rgb,(512, 349))
bigger = cv2.resize(img_rgb,(1980, 1048))
shrinking = cv2.resize(img_rgb, (1600, 800), interpolation=cv2.INTER_AREA)

Title = ['Original', 'Half', 'Bigger','Shrinked']
images = [img_rgb, half, bigger, shrinking]
count = 4

# Display all the images
for i in range(count):
    plt.subplot(2,2,i+1)
    plt.title(Title[i])
    plt.imshow(images[i])
    plt.axis('off')

plt.show()

```

## Task 3

```python
if img is not None:
    # Constant for the formula
    a = 5.0
    b = 30
    
    # Adjust brightness and contrast (alpha > 1.0 => Increase contrast, alpha < 1.0 => Decrease contrast) (beta > 0 => increase brightness and backward)
    
    img_processed = cv2.convertScaleAbs(img_rgb, alpha=a, beta=b)
    
    plt.figure(figsize=(12, 6))
    plt.subplot(1,2,1), plt.imshow(img_rgb), plt.title('Original Image'), plt.axis('off')
    plt.subplot(1,2,2), plt.imshow(img_processed), plt.title('Brightness Adjusted Image'), plt.axis('off')
    plt.show()
else:
    print("Failed to load the image. Please check the file path.")
```

## Task 4

### Creating Histogram for the given image

```python
if img is not None:
    color = ('r', 'g', 'b')
    # Calculate histogram of the image
    for i, col in enumerate(color):
        hist = cv2.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(hist, color=col)
        plt.xlim([0, 256])
        
    plt.title('Histogram for RGB Channels')
    plt.xlabel('Pixel Values')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Failed to load the image. Please check the file path.")
```

### Equalizing Image

```python
if img is not None:
    # Convert image from RGB to YUV 
    img_yuv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2YUV)
    
    # Equalize the histogram of Y channel (the luminance channel which is took care of Brightness)
    img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
    
    # Convert back to RGB
    img_equalized = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
    
    # Display original and equalized image
    plt.figure(figsize=(12, 6))
    plt.subplot(121), plt.imshow(img_rgb), plt.title('Original Image'), plt.axis('off')
    plt.subplot(122), plt.imshow(img_equalized), plt.title('Histogram Equalized'), plt.axis('off')
    plt.show()
else:
    print("Failed to load the image. Please check the file path.")
```

### Display the Histogram of Equalized Image

```python
if img is not None:
    color = ('r', 'g', 'b')
    # Calculate histogram of the image
    for i, col in enumerate(color):
        hist = cv2.calcHist([img_equalized], [i], None, [256], [0, 256])
        plt.plot(hist, color=col)
        plt.xlim([0, 256])
        
    plt.title('Histogram for RGB Channels of Equalized Image')
    plt.xlabel('Pixel Values')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Failed to load the image. Please check the file path.")
```

## Task 5

```python
if img is not None:
    # Convert the color image to Gray
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
    
    # Thresholding Technique
    ret, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)
    
    # Plotting the image
    plt.figure(figsize=(12,6))
    plt.subplot(121), plt.imshow(img_gray), plt.title('Gray Image'), plt.axis('off')
    plt.subplot(122), plt.imshow(thresh), plt.title('Thresholding Image'), plt.axis('off')
    plt.show()
else:
    print("Failed to load the image. Please check the file path.")
```
